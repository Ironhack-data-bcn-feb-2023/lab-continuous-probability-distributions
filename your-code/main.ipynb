## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here

def function(bottom,ceiling, count): 
    
    x = uniform.rvs(size=count)
    func_ = bottom + (ceiling - bottom)*x 
    
    return func_   
```

```python
uniform1 = function(10,15,100)
```

```python
uniform2 = function (10,60,1000)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, (ax1,ax2) = plt.subplots(ncols=2, sharey = True, figsize=(15,10))

sns.histplot(x=uniform1, bins = 10, ax=ax1);

sns.histplot(x=uniform2, bins=10, ax=ax2);

ylim = (0, 130)
ax1.set_ylim(ylim)
ax2.set_ylim(ylim)
```

How are the two distributions different?

```python
# your answer below

"""En primer lugar, el rango de las dos distribuciones uniformes es diferente dado que el valor máximo de la 
segunda distribución es mayor, por lo que la probabilidad será menor. Por otro lado, el tamaño de la muestra es diferente,
la muestra de la segunda distribución es más grande que la primera, por lo que la media y la varianza de la segunda
distribución seran más precisos como estimadores de la población de dicha muestra"""
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm
```

```python
def function_normal (average, st_dev, count):
    
    dist_norm = norm(average,st_dev)
    muestra = dist_norm.rvs(count)
    
    return muestra 
```

```python
normal1 = function_normal (10,1,1000)
normal2 = function_normal (10,50,1000)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, sharey = True, figsize=(15,10))

sns.histplot(x=normal1, ax=ax1);

sns.histplot(x=normal2, ax=ax2);

ylim = (0, 120)
ax1.set_ylim(ylim)
ax2.set_ylim(ylim)
```

How are the two distributions different?

```python
# your answer below

"""La desviación estándard indica la cantidad de dispersión de un conjunto de datos, es decir, como mayor sea la desviación
estandard, los datos estan más alejados de la media. En el caso de estas dos distribuciones, hay una mayor variación en 
la segundadistribución dado que tiene una st_Dev de 50, lo que significa que los valores pueden variar de +,-50 respecto la media"""
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 2 and 200. 

1. Use the function to generate two number sequences with the size of 1000 for both.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
import numpy as np
```

```python
# your code here

def function_exp (mean):
    
    dist_exp = np.random.exponential (mean,1000)
    
    return dist_exp
```

```python
exp1 = function_exp(2)
exp2 = function_exp (200)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, sharey = True, figsize=(15,10))

sns.histplot(x=exp1, ax=ax1, bins = 100);

sns.histplot(x=exp2, ax=ax2, bins = 100);

ylim = (0, 80)
ax1.set_ylim(ylim)
ax2.set_ylim(ylim)
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
from scipy.stats import expon
```

```python
# your answer here
# Hint: This is same as saying P(x<15)

time_bank = expon(10)
prob_less15 = time_bank.cdf(14)
prob_less15
```

What is the probability that the customer will spend more than 15 minutes

```python
# your answer here
prob_more15 = 1 - prob_less15
prob_more15
```
