## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def generate_uniform_random_numbers(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    random_numbers = bottom + (ceiling - bottom) * x
    return random_numbers

# Parameters for the first set
bottom1 = 10
ceiling1 = 15
count1 = 100

# Parameters for the second set
bottom2 = 10
ceiling2 = 60
count2 = 1000
# Generate normally distributed random numbers
random_numbers_set1 = generate_uniform_random_numbers(bottom1, ceiling1, count1)
random_numbers_set2 = generate_uniform_random_numbers(bottom2, ceiling2, count2)

# Plot histograms
plt.figure(figsize=(12, 6))

# Histogram for the first set
plt.subplot(1, 2, 1)
plt.hist(random_numbers_set1, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution (bottom=10, ceiling=15, count=100)')
plt.grid(True)

# Histogram for the second set
plt.subplot(1, 2, 2)
plt.hist(random_numbers_set2, bins=10, color='lightgreen', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution (bottom=10, ceiling=60, count=1000)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt

def generate_normal_random_numbers(mean, std_dev, count):
    return np.random.normal(mean, std_dev, count)

# Parameters for the first set
mean1 = 10
std_dev1 = 1
count1 = 1000

# Parameters for the second set
mean2 = 10
std_dev2 = 50
count2 = 1000

# Generate normally distributed random numbers
random_numbers_set1 = generate_normal_random_numbers(mean1, std_dev1, count1)
random_numbers_set2 = generate_normal_random_numbers(mean2, std_dev2, count2)

#histograms
plt.figure(figsize=(12, 6))

# Histogram for the first set
plt.subplot(1, 2, 1)
plt.hist(random_numbers_set1, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution (mean=10, std_dev=1, count=1000)')
plt.grid(True)

# Histogram for the second set
plt.subplot(1, 2, 2)
plt.hist(random_numbers_set2, bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution (mean=10, std_dev=50, count=1000)')
plt.grid(True)

plt.tight_layout()
plt.show()


```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 2 and 200. 

1. Use the function to generate two number sequences with the size of 1000 for both.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt

def generate_exponential_random_numbers(mean, size):
    return np.random.exponential(scale=mean, size=size)

# Parameters for the first sequence
mean1 = 2
size1 = 1000

# Parameters for the second sequence
mean2 = 200
size2 = 1000

# Generate exponentially distributed random numbers
random_numbers_seq1 = generate_exponential_random_numbers(mean1, size1)
random_numbers_seq2 = generate_exponential_random_numbers(mean2, size2)

# Plot
plt.figure(figsize=(12, 6))

# Histogram for the first sequence
plt.subplot(1, 2, 1)
plt.hist(random_numbers_seq1, bins=100, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Exponential Distribution (Mean = 2, Size = 1000)')
plt.grid(True)

# Histogram for the second sequence
plt.subplot(1, 2, 2)
plt.hist(random_numbers_seq2, bins=100, color='lightgreen', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Exponential Distribution (Mean = 200, Size = 1000)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your answer here
# Hint: This is same as saying P(x<15)
import math

#exponential distribution (in minutes)
mean_minutes = 10

# Convert mean to rate parameter (λ)
rate_parameter = 1 / mean_minutes

# Time in minutes for which we want to calculate the probability
time_minutes = 15

# Calculate the probability
probability_less_than_15_minutes = 1 - math.exp(-rate_parameter * time_minutes)

print(f"The probability that a customer will spend less than fifteen minutes in the bank is: {probability_less_than_15_minutes:.3f}")
```
