## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_uniform_randoms(bottom, ceiling, count):
    x = np.random.uniform(size=count)
    randoms = bottom + (ceiling - bottom) * x
    return randoms

randoms_set1 = generate_uniform_randoms(bottom=10, ceiling=15, count=100)

randoms_set2 = generate_uniform_randoms(bottom=10, ceiling=60, count=1000)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(randoms_set1, bins=10, color='blue')
plt.title('Uniform Distribution (bottom=10, ceiling=15)')
plt.xlabel('Value')
plt.ylabel('Count')


plt.subplot(1, 2, 2)
plt.hist(randoms_set2, bins=10, color='blue')
plt.title('Uniform Distribution (bottom=10, ceiling=60)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.show()
```

How are the two distributions different?

The larger sample size results in a more accurate representation of the uniform distribution.

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_normal_randoms(mean, std_dev, count):
    randoms = np.random.normal(loc=mean, scale=std_dev, size=count)
    return randoms

mean1 = 10
std_dev1 = 1
count1 = 1000
randoms_set1 = generate_normal_randoms(mean1, std_dev1, count1)

mean2 = 10
std_dev2 = 50
count2 = 1000
randoms_set2 = generate_normal_randoms(mean2, std_dev2, count2)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(randoms_set1, bins=30, color='blue')
plt.title('Normal Distribution (mean=10, std_dev=1)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(randoms_set2, bins=30, color='blue')
plt.title('Normal Distribution (mean=10, std_dev=50)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.show()
```

How are the two distributions different?

- Plot 1 has a small standard deviation, resulting in a narrower peak around the mean value (10).
- Plot 2 has a large standard deviation, resulting in a broader shape, indicating a wider spread of data points across a larger range of values.

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 2 and 200. 

1. Use the function to generate two number sequences with the size of 1000 for both.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
import numpy as np

def generate_exponential_numbers(mean, size):
    scale = mean
    return np.random.exponential(scale=scale, size=size)

mean1 = 2
mean2 = 200
size = 1000

sequence1 = generate_exponential_numbers(mean1, size)
sequence2 = generate_exponential_numbers(mean2, size)

plt.subplot(1, 2, 1)
plt.hist(sequence1, bins=100, color='blue')
plt.title('Exponential Distribution (mean=2)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(sequence2, bins=100, color='blue')
plt.title('Exponential Distribution (mean=200)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.show()
```

How are the two distributions different?

the primary difference between the two distributions is their central tendency, which is determined by the mean parameter. The distribution with a lower mean (mean=2) is skewed towards smaller values, while the distribution with a higher mean (mean=200) is skewed towards larger values. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
import numpy as np

def exponential_cdf(x, mean):
    rate = 1 / mean
    cdf = 1 - np.exp(-rate * x)
    return cdf

mean_minutes = 10

time_minutes = 15

probability_less_than_15_minutes = exponential_cdf(time_minutes, mean_minutes)*100

print(f"The probability that a customer will spend less than fifteen minutes in the bank is: {probability_less_than_15_minutes:.2f}%")
```

What is the probability that the customer will spend more than 15 minutes

```python
probability_more_than_15_minutes = 100 - probability_less_than_15_minutes

print(f"The probability that a customer will spend more than fifteen minutes in the bank is: {probability_more_than_15_minutes:.2f}%")
```
