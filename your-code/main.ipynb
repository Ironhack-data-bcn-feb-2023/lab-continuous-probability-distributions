**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generate_uniform_random_numbers(bottom, ceiling, count):
    x = stats.uniform.rvs(size=count)
    randoms = bottom + (ceiling - bottom) * x
    return randoms

random_numbers_1 = generate_uniform_random_numbers(bottom=10, ceiling=15, count=100)

random_numbers_2 = generate_uniform_random_numbers(bottom=10, ceiling=60, count=1000)

plt.hist(random_numbers_1, bins=10, alpha=0.5, label='bottom=10, ceiling=15, count=100')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution - bottom=10, ceiling=15, count=100')
plt.legend()
plt.show()

plt.hist(random_numbers_2, bins=10, alpha=0.5, label='bottom=10, ceiling=60, count=1000')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution - bottom=10, ceiling=60, count=1000')
plt.legend()
plt.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
def generate_normal_random_numbers(mean, std_dev, count):
    randoms = np.random.normal(loc=mean, scale=std_dev, size=count)
    return randoms

random_numbers_1 = generate_normal_random_numbers(mean=10, std_dev=1, count=1000)

random_numbers_2 = generate_normal_random_numbers(mean=10, std_dev=50, count=1000)

plt.hist(random_numbers_1, bins=50, alpha=0.5, label='mean=10, std_dev=1')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution - mean=10, std_dev=1')
plt.legend()
plt.show()

# Plot the second histogram
plt.hist(random_numbers_2, bins=50, alpha=0.5, label='mean=10, std_dev=50')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution - mean=10, std_dev=50')
plt.legend()

plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 2 and 200. 

1. Use the function to generate two number sequences with the size of 1000 for both.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
def generate_exponential_numbers(mean, size):
    randoms = np.random.exponential(scale=mean, size=size)
    return randoms

sequence1 = generate_exponential_numbers(mean=2, size=1000)
sequence2 = generate_exponential_numbers(mean=200, size=1000)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

axes[0].hist(sequence1, bins=100, alpha=0.5)
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Count')
axes[0].set_title('Exponential Distribution - Mean = 2')

axes[1].hist(sequence2, bins=100, alpha=0.5)
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Count')
axes[1].set_title('Exponential Distribution - Mean = 200')

plt.tight_layout() 
plt.show()
```

How are the two distributions different?

the graphical representation of the distributions is very similar in both cases,but the values in the first graphic are concentrated from 0 to 10  

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
def exponential_cdf(x, mean):
    rate_parameter = 1 / mean
    
    probability = 1 - np.exp(-rate_parameter * x)
    
    return probability

mean = 10

time_in_bank = 15
probability_less_than_15 = exponential_cdf(time_in_bank, mean)

print(f"The probability that a customer will spend less than 15 minutes in the bank is: {probability_less_than_15:.3f}")
```

What is the probability that the customer will spend more than 15 minutes

```python
probability_more_than_15 = 1 - probability_less_than_15

print(f"The probability that a customer will spend more than 15 minutes in the bank is: {probability_more_than_15:.3f}")
```
